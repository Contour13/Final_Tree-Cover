### Plotting Single Variable Distributions

```python
for col in num_cols:
    sns.kdeplot(soilsCovDF[col])
    plt.show()
```

```python
cat_cols = [
    "Cover_Type",
    "GeoSoils",
    "Wilderness_Area1",
    "Wilderness_Area2",
    "Wilderness_Area3", 
    "Wilderness_Area4" 
]

for col in cat_cols:
    sns.countplot(col, data=soilsCovDF)
    plt.show()
```

GeoSoils are highly rocky due to igneous/metamorphic category. Cover Type is largely attributed to be Lodgepole Pine and Spruce/Fir in Wilderness areas 1 and 3.

### Univariate and multi-variate relationships

```python
sns.regplot(x="Elevation", y="Cover_Type", data=soilsCovDF, fit_reg=False, y_jitter=0.1)
```

```python
sns.regplot(x="Elevation", y="Wilderness_Area1", data=soilsCovDF, fit_reg=False, y_jitter=0.1)
```

Wilderness Area 1 appears to have occurences of recorded trees at 2500 meters and above in elevation up to 3700m.

```python
sns.regplot(x="Elevation", y="Wilderness_Area2", data=soilsCovDF, fit_reg=False, y_jitter=0.1)
```

Wilderness Area 2 appears to have occurences of recorded trees at 2900 meters and above in elevation up to 3600m.

```python
sns.regplot(x="Elevation", y="Wilderness_Area3", data=soilsCovDF, fit_reg=False, y_jitter=0.1)
```

Wilderness Area 3 has an interesting split in dataset around 3600 meters. Entire range of trees vary in elevation from 2250 to 3900 meters.

```python
sns.regplot(x="Elevation", y="Wilderness_Area4", data=soilsCovDF, fit_reg=False, y_jitter=0.1)
```

Wilderness Area 4 has recorded trees in low elevations from approximately 1900 meters to 2600 meters. This would suggest that this area is lowest in elevation compared to the other areas studies.

```python
sns.regplot(x="Horizontal_Distance_To_Hydrology", y="Cover_Type", data=soilsCovDF, fit_reg=False, y_jitter=0.1)
```

The lower distances to hydrology suggests that these species are likely in areas that are more wet and/or lower in elevation (i.e. near creeks, river valleys, etc.)

```python
def makeUniPlot(xinput):
    sns.regplot(x=xinput, y="Cover_Type", data=soilsCovDF, fit_reg=False, y_jitter=0.1, x_jitter=0.1, scatter_kws={"alpha": 0.2})
    plt.show()
```

```python
for col in num_cols:
    makeUniPlot(col)
```

I don't think I need roadways category as I feel this is unnecessary for analysis. lower elevations are going to have more roads due to it being easier to build roads and pathas compared to higher elevations where severe grading has to occur, etc.

```python
def makeUniPlot(xinput):
    sns.regplot(x=xinput, y="GeoSoils", data=soilsCovDF, fit_reg=False, y_jitter=0.1, x_jitter=0.1, scatter_kws={"alpha": 0.2})
    plt.show()
```

```python
for col in num_cols:
    makeUniPlot(col)
```
